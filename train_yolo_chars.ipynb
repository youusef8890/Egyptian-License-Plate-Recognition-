```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(
    data='/content/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    name='yolo_egypt_chars',
    project='ocr_project'
)
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# ğŸ”§ Load model
model = YOLO("/content/ocr_project/yolo_egypt_chars/weights/best.pt")

# ğŸ” Load image (plate image)
img_path = "/content/License_plate_Egypt_private.jpg"  # Replace with any test plate
results = model(img_path, conf=0.25)  # You can adjust confidence threshold

# ğŸ“¦ Extract result
res_plotted = results[0].plot()

# ğŸ–¼ï¸ Show detection
plt.imshow(res_plotted[..., ::-1])
plt.title("Detected Characters")
plt.axis('off')
plt.show()
```
